```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of points
n_camera_bins = 11
points = np.arange(n_camera_bins)
```

```python
points_normalized = points - n_camera_bins // 2
```

```python
print(points_normalized)
print(n_camera_bins)
```

```python
np.abs(points_normalized) / (n_camera_bins - 1) * 2
```

```python

# Vectorized mapping function
points_mapped = (11 ** (np.abs(points_normalized) / (n_camera_bins - 1) * 2) - 1) * np.sign(points_normalized)
```

```python

# Plotting
plt.figure(figsize=(6, 6))
plt.scatter(points_mapped, points, s=30, color='royalblue', label='Mapped Points')
plt.xlabel('Mapped Value')
plt.ylabel('Original Value')
plt.title('Mapping of Points with Custom Function')
plt.gca().set_aspect('equal')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```
